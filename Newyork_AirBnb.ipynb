# **UNDERSTAND THE GIVEN VARIABLES**
**id :-** This is a unique identifier for each listing in the dataset.

**name :-** This is the name or title of the listing, as it appears on the Airbnb website.

**Host_id :-** This is a unique identifier for each host in the dataset.

**Host_name :-** This is the name of the host as it appears on the Airbnb website.

**Neighbourhood_group :-** This is a grouping of neighborhoods in New York City, such as Manhattan or Brooklyn.

**Neighbourhood :-** This is the specific neighborhood in which the listing is located.

**Latitude :-** This is the geographic latitude of the listing.

**Longitude :-** This is the geographic longitude of the listing.

**Room_type :-** This is the type of room or property being offered, such as an entire home, private room, shared room.

**Price :-** This is the nightly price for the listing, in US dollars.

**Minimum_nights :-** This is the minimum number of nights that a guest must stay at the listing.

**number_of_reviews :-** This is the total number of reviews that the listing has received.

**last_review :-** This is the date on which last review added.

**Reviews_per_month :-** This is the average number of reviews that the listing receives per month.

**calculated_host_listings_count :-** This is the total number of listings that the host has on Airbnb.

**availability_365 :-** This is the number of days in the next 365 days that the listing is available for booking.

**number_of_reviews_ltm :-** Number of reviews in last 12 Months.

**rating :-** Number of rating given by travellers.

**bedrooms :-**  It may denote the number of bedrooms in the accommodation, typically represented by discrete numerical values.

**beds :-** This variable could represent the number of beds available in the accommodation, counted as discrete units.

**baths :-** This variable could represent the number of bathrooms in the listing, typically counted in discrete units.

```python
# idenfying outliers in price

df = data[data['price'] < 1500]

sns.boxplot(data=df, x='price')
```

 #### Distribution Of Airbnb Bookings Price Range Using Histogram

```python
# Price Distribution
# Create a figure with a custom size
plt.figure(figsize=(12, 5))

# Set the seaborn theme to darkgrid
sns.set_theme(style='darkgrid')

# Create a histogram of the 'price' column of the Airbnb_df dataframe
# using sns distplot function and specifying the color as red
sns.distplot(df['price'],color=('r'))

# Add labels to the x-axis and y-axis
plt.xlabel('Price', fontsize=14)
plt.ylabel('Density', fontsize=14)

# Add a title to the plot
plt.title('Distribution of Airbnb Prices',fontsize=15)
```

**INSIGHTS -->**
  
  * The range of prices being charged on Airbnb appears to be from 10 to 1500 dollars , with the majority of listings falling in the price range of 50       to 200 dollars.
  * The distribution of prices appears to have a peak in the 50 to 150 dollars range, with a relatively lower density of listings in higher and lower        price ranges.

## Number of Max. Reviews by Each Neighborhood Group using Pie Chart

```python
# Group the Airbnb data by neighbourhood group
reviews_by_neighbourhood_group = df.groupby("neighbourhood_group")["number_of_reviews"].max()

# Create a pie chart to visualize the distribution of maximum number of reviews among different neighbourhood groups
plt.pie(reviews_by_neighbourhood_group, labels=reviews_by_neighbourhood_group.index, autopct='%1.1f%%')

# Add a title to the chart
plt.title("Number of maximum Reviews by Neighborhood Group in NYC", fontsize='15')

# Display the chart
plt.show()
```

**Observation -->**
* Manhattan and Queens seem to be the most popular neighborhoods for reviewing, as they have both high number of maximum reviews.
* This data suggests that  Manhattan,Queens, and Brooklyn are the most popular neighborhoods for tourists or visitors, based on the high number of reviews they receive.

## Correlation Heatmap Visualization

```python
# heat map - correlation of one variable with others for numerical column

corr = df[['id','host_id', 'latitude','longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count','availability_365']].corr()
corr

# plt.figure(figsize=(8, 6))
# sns.heatmap(data=corr, annot=True)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(data=corr, annot=True)
```

**Observations -->**

* There is a weak negative correlation (-0.016) between the price column and the calculated_host_listings_count column, which suggests that hosts with more listings tend to charge higher prices for their listings.
* There is a moderate positive correlation (0.23) between the calculated_host_listings_count column and the availability_365 column, which suggests that hosts with more listings tend to have more days of availability in the next 365 days.
* There is a moderate positive correlation (0.42) between the host_id and id columns, which suggests that hosts with more listings are more likely to      have unique host IDs.s.
